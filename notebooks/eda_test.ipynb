## Salary Insights

- Salary vs Job Type: Average or median salary grouped by job type.

- Salary by Company: Identify companies offering the highest median salaries.

- Salary Distribution: Histogram or boxplot (after cleaning and parsing salary values).

```python
df
```

```python
df_salary_clean = df.dropna(subset=['Salary']).copy()
```

```python
import re

def extract_salary_range(s):
    if pd.isna(s):
        return pd.Series([None, None])

    s_clean = s.replace('$', '').replace(',', '').lower()
    match = re.findall(r'(\d+(?:\.\d+)?)([kKmM]?)', s_clean)

    if not match:
        return pd.Series([None, None])

    values = []
    for val, suffix in match:
        num = float(val)
        if suffix == 'k':
            num *= 1_000
        elif suffix == 'm':
            num *= 1_000_000
        values.append(num)

    if len(values) == 1:
        return pd.Series([values[0], values[0]])
    elif len(values) >= 2:
        return pd.Series([min(values[:2]), max(values[:2])])
    else:
        return pd.Series([None, None])
```

```python
df_salary_clean[['Min_Salary', 'Max_Salary']] = df_salary_clean['Salary'].apply(extract_salary_range)
df_salary_clean
```

```python
def normalize_salary(value):
    if pd.isna(value):
        return None
    try:
        # Count number of digits before the decimal point
        digits = len(str(int(value)))
        if digits <= 3:
            return value * 2080  # Convert hourly to annual
        return value  # Already annual
    except:
        return None
```

```python
df_salary_clean['Min_Salary'] = df_salary_clean['Min_Salary'].apply(normalize_salary)
df_salary_clean['Max_Salary'] = df_salary_clean['Max_Salary'].apply(normalize_salary)
```

```python
df_salary_clean['Avg_Salary'] = df_salary_clean[['Min_Salary', 'Max_Salary']].mean(axis=1)
```

```python
df_salary_clean = df_salary_clean[df_salary_clean['Avg_Salary'] != 0]
```

```python
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from wordcloud import WordCloud
```

```python
# Step 1: Combine and clean descriptions
text = ' '.join(df['Description'].dropna()).lower()
text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation/numbers
```

```python
# Step 2: Tokenize and tag
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
```

```python
action_verbs = [word for word, tag in tagged if tag == 'VB' and word not in stopwords.words('english')]
```

```python
# Step 4: Count the most common verbs
verb_counts = Counter(action_verbs)
top_verbs = verb_counts.most_common(20)
```

```python
print("Top action verbs in job descriptions:")
for verb, count in verb_counts.most_common(20):
    print(f"{verb}: {count}")
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(verb_counts)
```

```python
# Plot the word cloud
plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Action Verbs in Job Descriptions')
plt.show()

top_verbs
```

```python
# Define soft skill and responsibility keywords
SOFT_SKILL_PHRASES = {
    'communication': ['communication', 'communicate effectively', 'verbal and written'],
    'leadership': ['leadership', 'lead teams', 'manage teams'],
    'teamwork': ['teamwork', 'collaboration', 'work with cross-functional teams'],
    'problem-solving': ['problem solving', 'analytical skills', 'solve problems'],
    'time management': ['time management', 'prioritize tasks'],
    'adaptability': ['adaptability', 'adapt to change'],
    'critical thinking': ['critical thinking', 'strategic thinking'],
    'interpersonal skills': ['interpersonal skills', 'people skills'],
    'customer focus': ['customer service', 'client interaction', 'customer support']
}
```

```python
# Initialize counts
soft_skill_counts = {}

# Lowercase the descriptions for consistency
df['Description_clean'] = df['Description'].str.lower().fillna("")
```

```python
# Count matches
for skill, phrases in SOFT_SKILL_PHRASES.items():
    pattern = '|'.join([re.escape(p) for p in phrases])
    soft_skill_counts[skill] = df['Description_clean'].str.contains(pattern, regex=True).sum()
```

```python
# Convert to DataFrame
soft_skills_df = pd.DataFrame(soft_skill_counts.items(), columns=['Soft Skill / Responsibility', 'Mentions'])
soft_skills_df = soft_skills_df.sort_values(by='Mentions', ascending=False).reset_index(drop=True)
```

```python
soft_skills_df
```

```python
# Plot bar chart
plt.figure(figsize=(10, 6))
soft_skills_df_sorted = soft_skills_df.sort_values(by='Mentions', ascending=True)
plt.barh(soft_skills_df_sorted['Soft Skill / Responsibility'], soft_skills_df_sorted['Mentions'])
plt.xlabel('Mentions')
plt.title('Top Soft Skills and Responsibilities in Job Descriptions')
plt.tight_layout()
plt.show()
```

```python
# Plot word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(
    dict(zip(soft_skills_df['Soft Skill / Responsibility'], soft_skills_df['Mentions']))
)

plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Soft Skills and Responsibilities')
plt.show()
```
